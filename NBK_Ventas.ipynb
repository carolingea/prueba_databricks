```python
%sql

DROP TABLE  IF EXISTS ventas;

CREATE TABLE IF NOT EXISTS ventas as 
SELECT
    ROW_NUMBER() OVER (ORDER BY productid) AS Id,
    productid,
    customerid,
    Price,
    Quantity,
    Price * Quantity AS Total
FROM (
    SELECT
        productid,
        customerid,
        numinterno,
        Price,
        CAST(FLOOR(RAND() * 5) + 1 AS INT) AS Quantity
    FROM (
        SELECT DISTINCT
            p.`index` as ProductId,
            c.`Index` AS CustomerId,
            p.`Internal ID` as NumInterno,
            p.Price
        FROM products p
        LEFT JOIN customers c ON p.Index = c.Index
        ORDER BY RAND()
    )
) as t


```
